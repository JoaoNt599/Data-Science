# Calculando IQR para remover os outliers

```python
# Estatísticas descritivas
q1 = df["valor_aluguel"].quantile(0.25)
q3 = df["valor_aluguel"].quantile(0.75)
iqr = q3 - q1
```

```python
# Limites inferior e superior
limite_inferior = q1 - 1.5 * iqr
limite_superior = q3 + 1.5 * iqr
```

```python
# Removendo os outliers
df_sem_outliers = df[(df["valor_aluguel"] >= limite_inferior) & (df["valor_aluguel"] <= limite_superior)]

print(f"Número de registros antes: {len(df)}")
print(f"Número de registros depos: {len(df_sem_outliers)}")
```

```python
plt.figure(figsize=(10, 5))

# Antes
plt.subplot(1, 2, 1)
sns.boxplot(x=df["valor_aluguel"])
plt.title("Com Outliers")

# Depois
plt.subplot(1, 2, 2)
sns.boxplot(x=df_sem_outliers["valor_aluguel"])
plt.title("Sem Outliers")

plt.tight_layout()
plt.show()
```

# Regressão Linear Simples com Modelo OLS(Ordinary Least Squares)

```python
# Variável dependente
y = df["valor_aluguel"]

# Variável independente
X = df["area_m2"]
```

```python
# O Statsmodels requer a adicção de uma constante à variável independente
X = sm.add_constant(X)
```

```python
# Modelo criado
modelo = sm.OLS(y, X)
```

```python
# Treinamento do modelo
resultado = modelo.fit()
```

```python
print(resultado.summary())
```

# Plot para previsões do Modelo

```python
plt.figure(figsize = (12, 8))
plt.xlabel("area_m2", size = 16)
plt.ylabel("valor_aluguel", size = 16)
plt.plot(X["area_m2"], y, "o", label = "Dados Reais")
plt.plot(X["area_m2"], resultado.fittedvalues, "r-", label = "Linha de Regressão (Previsão do Modelo)")
plt.legend(loc = "best")
plt.show()
```
